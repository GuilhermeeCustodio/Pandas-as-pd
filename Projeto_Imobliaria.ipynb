## Analise Exploratória de dados

#### Qual o valor médio de aluguel por tipo de imóvel

```python
dados.head()
```

```python
dados["Valor"].mean()
```

```python
dados.groupby("Tipo").mean(numeric_only=True)
```

```python
dados.groupby("Tipo")["Valor"].mean()
```

```python
# Colchetes duplos, transforma em um dataframe
dados.groupby("Tipo")[["Valor"]].mean()
```

```python
# Sort_Values cria uma lista ordenada do menor para o maior
dados.groupby("Tipo")[["Valor"]].mean().sort_values("Valor")
```

```python
df_preco_tipo = dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')

df_preco_tipo.plot(kind='barh', figsize=(14, 10)), color ='purple';
```

## Removendo os imóveis comerciais

```python
dados.Tipo.unique()
```

```python
imoveis_comerciais = ['Conjunto Comercial/Sala','Prédio Inteiro','Loja/Salão','Galpão/Depósito/Armazém','Casa Comercial','Terreno Padrão','Loja Shopping/ Ct Comercial','Box/Garagem','Chácara','Loteamento/Condomínio','Sítio', 'Pousada/Chalé','Hotel','Industria']
```

```python
dados.query('Tipo in @imoveis_comerciais')
```

```python
dados.query('Tipo not in @imoveis_comerciais')
```

```python
df = dados.query('Tipo not in @imoveis_comerciais')
df.head()
```

```python
df.Tipo.unique()
```

```python
df_preco_tipo = df.groupby("Tipo")[["Valor"]].mean().sort_values("Valor")

df_preco_tipo.plot(kind='barh', figsize=(14, 10), color ='purple');
```
